```python
#Outlier detection
collist=data.columns.values
ncol=18
nrow=16
plt.figure(figsize=(ncol,2*ncol))
for i in range(1,len(collist)):
    plt.subplot(nrow,ncol,i+1)
    sns.boxplot(data[collist[i]],color='blue',orient='v')
    
```

```python
# As per the above box plot, outliers are present in the data, hence removing using zscore
from scipy.stats import zscore
z_sc=np.abs(zscore(data))
print(data.shape)
data=data.loc[(z_sc<3).all(axis=1)]
print(data.shape)
```

```python
y=data.iloc[:,0:1]
y
```

```python
x=data.iloc[:,1:]
x
```

```python
#Scaling the attributes
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
def r2scoref(reg,x,y):
    max_r_score=0
    for r in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r,test_size=0.30)
        reg.fit(x_train,y_train)
        y_pred=reg.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            r_state=r
            plt.scatter(x=y_test,y=y_pred)
    print('Maximum r2 score corresponding to random state',r_state,'is',max_r_score)
    return(r_state)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
r2scoref(lr,x,y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor(n_neighbors=3)
r_state=r2scoref(knr,x,y)
```

```python
from sklearn.linear_model import Lasso
lso=Lasso(alpha=0.001)
r_state=r2scoref(lso,x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,max_depth=6, min_samples_leaf=50,n_jobs=4)
r_state=r2scoref(rf,x,y)
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor()
r_state=r2scoref(ada,x,y)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
r_state=r2scoref(gbr,x,y)
```

```python
from sklearn.model_selection import cross_val_score
print('Mean r2_score for Random Forest regression is:',cross_val_score(rf,x,y,cv=10,scoring="r2").mean())
print('Standard Deviation for Random Forest regression is :',cross_val_score(rf,x,y,cv=5,scoring="r2").std())
```
