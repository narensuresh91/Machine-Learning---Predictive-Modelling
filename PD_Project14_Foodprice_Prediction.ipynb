```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#import the csv file into pandas dataframe
train=pd.read_excel('Food_Data_Train.xlsx')
test=pd.read_excel('Food_Data_Test.xlsx')
```

```python
#Combining the train and test data into a single file
train['Source']='train'
test['Source']='test'
data=pd.concat([train,test],ignore_index=True)
print(train.shape,test.shape,data.shape)
```

```python
data
```

```python
data['LOCALITY'].value_counts()
```

```python
data.drop(['LOCALITY'],axis=1,inplace=True)
```

```python
data['TITLE'].value_counts()
```

```python
data['TITLE'].unique()
```

```python
data['VOTES'] = data['VOTES'].apply(lambda x: str(x)[0:-5])
```

```python
data
```

```python
data.drop(['TITLE','RESTAURANT_ID','TIME'],axis=1,inplace=True)
```

```python
data
```

```python
data=pd.get_dummies(data)
```

```python
data
```

```python
x_train=data.iloc[:,1:][data['Source_train']==1]
```

```python
x_train.shape
```

```python
y_train=data.iloc[:,0:1][data['Source_train']==1]
```

```python
y_train
```

```python
x_test=data.iloc[:,1:][data['Source_test']==1]
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
dtrain_predictions = model.predict(x_test)
```

```python
dtrain_predictions
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
a = cross_val_score(model, x_train, y_train, cv=5, scoring='neg_mean_squared_error')
```
