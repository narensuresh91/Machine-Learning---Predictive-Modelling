```python
#Outlier detection
collist=data.columns.values
ncol=21
nrow=8
plt.figure(figsize=(2*ncol,4*ncol))
for i in range(1,len(collist)):
    plt.subplot(nrow,ncol,i+1)
    sns.boxplot(data[collist[i]],color='blue',orient='v')
```

```python
#As per the above, certain outliers are present in the data,needs to be removed using zscore
from scipy.stats import zscore
z_sc=np.abs(zscore(data))
print(data.shape)
data=data.loc[(z_sc<3).all(axis=1)]
print(data.shape)
```

```python
data['Late']=data['Absenteeism time in hours'].apply(lambda x: 1 if x>=2 else 0)
```

```python
data.drop(['Absenteeism time in hours'],axis=1,inplace=True)
```

```python
#Splitting the attributes and target
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
y=y.values.reshape(-1,1)
```

```python
y.shape
```

```python
#Scaling the attributes
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x.shape
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

```python
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(data.corr(),annot=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,auc,roc_curve
```

```python
lr=LogisticRegression()
gnb=GaussianNB()
knc=KNeighborsClassifier(n_neighbors=8)
svc=SVC()
dtc=DecisionTreeClassifier(random_state=6)
rfc=RandomForestClassifier(n_estimators=200,random_state=42)
```

```python
models =[]
models.append(('LogisticRegression',lr))
models.append(('GaussianNB',gnb))
models.append(('KNeighborsClassifer',knc))
models.append(('SVC',svc))
models.append(('DecisionTreeClassifier',dtc))
models.append(('RandomForestClassifier',rfc))
```

```python
Model=[]
score=[]
cvs=[]
rocscore=[]
for name,model in models:
    print('**************',name,'***********')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pred=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pred)
    score.append(AS*100)
    print('Accuracy Score =',AS)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring='accuracy').mean()
    print('Cross Val Score =',sc)
    cvs.append(sc*100)
    false_positive_rate,true_positive_rate,thresholds = roc_curve(y_test,pred)
    roc_auc=auc(false_positive_rate,true_positive_rate)
    print('roc-auc score =',roc_auc)
    rocscore.append(roc_auc*100)
    print('\n')
    print('Classification Report\n',classification_report(y_test,pred))
    print('\n')
    print('Confusion Matrix \n',confusion_matrix(y_test,pred))
```

```python
f_result=pd.DataFrame({'Model' : Model, 'Accuracy Score': score,'Cross_Val_Score':cvs,'ROC-AUC_Curve':rocscore})
f_result
```
