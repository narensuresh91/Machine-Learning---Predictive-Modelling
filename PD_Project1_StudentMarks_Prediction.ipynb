```python
data['Percentage']=(data['Total']/800)*100
```

```python
data.drop(['Total'],axis=1,inplace=True)
```

```python
data.drop(['DOB'],axis=1,inplace=True)
```

```python
data.drop(['Name'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Gender']=le.fit_transform(data['Gender'])
```

```python
collist=data.columns.values
nrow=3
ncol=11
plt.figure(figsize=(ncol,10))
for i in range(1,len(collist)):
    plt.subplot(nrow,ncol,i+1)
    sns.boxplot(data[collist[i]],color='blue',orient='v')
    plt.tight_layout()
```

```python
from scipy.stats import zscore
z_sc=np.abs(zscore(data))
data=data[(z_sc<3).all(axis=1)]
```

```python
data
```

```python
data.skew()
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1:]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=43,test_size=0.2)
reg=LinearRegression()
reg.fit(x_train,y_train)
print(reg.score(x_train,y_train))
```

```python
y_pred=reg.predict(x_test)
y_pred
```

```python
y_test
```

```python
plt.scatter(x=y_test,y=y_pred)
```

```python
#The predicted values are in close proximity with the actual values, and hence the model is considered acceptable.
```
